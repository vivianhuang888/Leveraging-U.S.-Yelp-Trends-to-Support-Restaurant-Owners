## Cleaning Business Table




### Overview of current table

```python
%%bigquery
SELECT *
FROM `ba775-project-a04.yelp_data.business4`
LIMIT 10;
```

### Creating a new table filtering the businesses that are categorized as Restaurants or Food Establishments

```python
%%bigquery
CREATE OR REPLACE TABLE
  `ba775-project-a04.yelp_data.restaurant_business` AS

SELECT * EXCEPT(int64_field_0)
FROM
  `ba775-project-a04.yelp_data.business4`
WHERE
  categories LIKE '%Restaurant%'
    AND categories LIKE '%Food%'
    AND stars IS NOT NULL;
```

### Checking if any restaurants do not have a review

```python
%%bigquery
SELECT B.business_id, R.business_id, R.review_id
FROM `ba775-project-a04.yelp_data.restaurant_business` AS B
LEFT JOIN `ba775-project-a04.yelp_data.review` AS R
ON B.business_id = R.business_id
WHERE R.review_id IS NUll;
```

### <font face="Helvetica" size=4 color=2E0202> **Q1.2. How active are elite users compared to non-elite users in terms of fans, compliments, reviews, and contributions?** <font>

```python
%%bigquery
SELECT
   CASE WHEN elite IS NOT NULL THEN 'Elite' ELSE 'Non-Elite' END AS user_type,
   ROUND(AVG(review_count),2 ) AS avg_reviews,
   ROUND(AVG(useful), 2 ) AS avg_useful,
   ROUND(AVG(fans), 2) AS avg_fans,
   ROUND (AVG(
       compliment_hot + compliment_more + compliment_profile + compliment_cute +
       compliment_list + compliment_note + compliment_plain + compliment_cool +
       compliment_funny + compliment_writer + compliment_photos
   ), 2 ) AS avg_compliments,
   ROUND( AVG (average_stars), 2 ) AS avg_RatingOfAllReviews,
   ROUND (AVG(
       compliment_hot + compliment_more + compliment_profile + compliment_cute +
       compliment_list + compliment_note + compliment_plain + compliment_cool +
       compliment_funny + compliment_writer + compliment_photos + cool + useful + funny
   ), 2 )AS avg_interactions,
   COUNT(user_id) AS total_count
FROM `ba775-project-a04.yelp_data.restaurant_user`
GROUP BY user_type
ORDER BY user_type;
```

Elite users are recognized by Yelp for their high-quality contributions to the platform, including helpful reviews, useful tips, and active engagement with the community. These users stand out significantly compared to non-elite users in their activity and impact. On average, elite users write an impressive 396.79 reviews compared to just 34.37 by non-elite users. Their contributions are highly valued, as seen in the average of 1,611 useful votes per elite user. Additionally, elite users are popular within the Yelp community, boasting an average of 50 fans, which reflects the significance of their opinions. They also receive significantly more compliments on their reviews, showcasing their active participation and positive engagement. Interestingly, both elite and non-elite users exhibit relatively consistent rating behavior, with average ratings of 3.96 for elite users and 3.68 for non-elite users.

The higher engagement and impact of elite users can be attributed to their prominence on the platform. Yelp highlights their reviews, which increases visibility and attracts more interactions. Elite users are also known for their consistent quality, which further drives engagement from the community. Their elevated status as trusted contributors makes their reviews more influential, fostering a cycle of increased recognition and interaction that differentiates them from non-elite users.

To capitalize on this data, Yelp could take strategic steps to encourage further contributions from elite users while inspiring non-elite users to aspire to elite status. Recognizing elite users publicly for milestones and providing badges or levels could boost motivation and loyalty. Offering businesses a way to thank elite users for helpful reviews might also strengthen ties within the community. Finally, raising awareness among non-elite users about the benefits of elite status—such as increased visibility and influence—could encourage more participation and higher-quality contributions, enriching the overall Yelp experience.

### <font face="Helvetica" size=4 color=2E0202> **Q1.3. What are the most common compliment types for tips, and what do they reveal about user preferences?**<font>

```python
%%bigquery

WITH active_users AS (
    SELECT
        user_id
    FROM
        `ba775-project-a04.yelp_data.user`
    WHERE
        review_count > 100
),
filtered_users AS (
    SELECT
        user_id,
        name,
        compliment_list,
        compliment_hot,
        compliment_cute,
        compliment_more,
        compliment_profile,
        compliment_note,
        compliment_plain,
        compliment_cool,
        compliment_funny,
        compliment_photos,
        compliment_writer,
        fans
    FROM
        `ba775-project-a04.yelp_data.user`
    WHERE
        compliment_list > 1 -- at least one compliment
        AND user_id IN (SELECT user_id FROM active_users)
)
SELECT
    'Hot' AS compliment_type, SUM(compliment_hot) AS total_count FROM filtered_users
UNION ALL
SELECT
    'Cute', SUM(compliment_cute) FROM filtered_users
UNION ALL
SELECT
    'More', SUM(compliment_more) FROM filtered_users
UNION ALL
SELECT
    'Profile', SUM(compliment_profile) FROM filtered_users
UNION ALL
SELECT
    'Note', SUM(compliment_note) FROM filtered_users
UNION ALL
SELECT
    'Plain', SUM(compliment_plain) FROM filtered_users
UNION ALL
SELECT
    'Cool', SUM(compliment_cool) FROM filtered_users
UNION ALL
SELECT
    'Funny', SUM(compliment_funny) FROM filtered_users
UNION ALL
SELECT
    'Photos', SUM(compliment_photos) FROM filtered_users
UNION ALL
SELECT
    'Writer', SUM(compliment_writer) FROM filtered_users
ORDER BY
    total_count DESC;
```

The data reveals that customers most frequently give compliments in the categories "Plain" (3,723,478), "Cool" (3,541,876), and "Funny" (3,541,876), highlighting an appreciation for positive experiences, unique or trendy aspects, and humor. Categories like "Hot" (2,517,770) and "Photos" (1,602,189) also rank high, emphasizing the importance of visually appealing content and standout features. Less common categories such as "Note," "Writer," and "Cute" reflect more niche preferences, showing customer appreciation for thoughtful, creative, or personal elements.

For businesses, these trends underline the importance of delivering engaging and memorable experiences that resonate broadly with customers. The popularity of "Photos" points to the value of visual appeal, making it critical for businesses to prioritize aesthetics in food presentation, decor, and branding. Additionally, personalized touches like creative responses or thoughtful gestures can build deeper connections with customers, even if they appeal to a smaller audience.

To enhance engagement, Yelp could spotlight popular categories like "Plain," "Cool," and "Funny" through tags or highlights on business profiles. Encouraging photo uploads through contests or displays could further leverage the "Photos" category’s popularity. Providing businesses with insights into their most common compliments would help them better align their services with customer preferences, ultimately boosting loyalty and satisfaction.

### <font face="Helvetica" size=4 color=2E0202> **Q1.4. How does the number of friends influence star ratings and the number of compliments given?** <font>

```python
%%bigquery
WITH UserAggregates AS (
    SELECT
        u.user_id,
        COUNT(u.friends) AS no_of_friends,
        AVG(u.average_stars) AS avg_stars,
        COUNT(t.compliment_count) AS no_of_compliments
    FROM
        `ba775-project-a04.yelp_data.restaurant_user` AS u
    INNER JOIN
        `ba775-project-a04.yelp_data.restaurant_tip` AS t
    ON
        u.user_id = t.user_id
    GROUP BY
        u.user_id
),
BucketedData AS (
    SELECT
        user_id,
        no_of_friends,
        avg_stars,
        no_of_compliments,
        CASE
            WHEN no_of_friends BETWEEN 0 AND 30 THEN '0-30'
            WHEN no_of_friends BETWEEN 31 AND 80 THEN '31-80'
            WHEN no_of_friends BETWEEN 81 AND 120 THEN '81-120'
            WHEN no_of_friends BETWEEN 121 AND 160 THEN '121-160'
            WHEN no_of_friends BETWEEN 161 AND 200 THEN '161-200'
            WHEN no_of_friends BETWEEN 201 AND 250 THEN '201-250'
            WHEN no_of_friends BETWEEN 251 AND 300 THEN '251-300'
            WHEN no_of_friends BETWEEN 301 AND 350 THEN '301-350'
            WHEN no_of_friends BETWEEN 351 AND 400 THEN '351-400'
            WHEN no_of_friends BETWEEN 401 AND 450 THEN '401-450'
            WHEN no_of_friends BETWEEN 451 AND 500 THEN '451-500'
            WHEN no_of_friends BETWEEN 501 AND 550 THEN '501-550'
            ELSE '551 and over'
        END AS friend_bucket
    FROM
        UserAggregates
)
SELECT
    friend_bucket,
    AVG(avg_stars) AS avg_star_rating,
    AVG(CASE WHEN no_of_compliments >= 1 THEN no_of_compliments ELSE NULL END) AS avg_compliments
FROM
    BucketedData
GROUP BY
    friend_bucket
ORDER BY
    avg_compliments DESC;
```

In this query, users were categorized into buckets based on their number of friends, ranging from 0-30 to 551+ friends, to analyze the impact on review ratings and engagement (measured by the average number of compliments). The output reveals a clear trend: as the number of friends increases, the likelihood of receiving compliments rises significantly. This is evident from the drastic jumps in engagement, such as an increase of 45 compliments between the 0-30 and 31-80 friend buckets, and an impressive rise of 2,491 compliments between the 501-550 and 551+ friend buckets.

This suggests that users with larger friend networks on Yelp experience higher engagement, possibly because their reviews are more frequently featured in recommended sections or gain greater visibility within the community. Additionally, these users may put more effort into their reviews, offering detailed and thoughtful feedback about their experiences, which resonates with other users who then leave compliments. Building a strong friend network on Yelp may also reflect a user’s dedication to engaging meaningfully with the platform, further amplifying their influence.

For businesses, engaging with these socially connected users is a strategic opportunity to boost visibility and credibility. Personalized interactions, such as thanking them for reviews, addressing their feedback thoughtfully, or offering exclusive perks like discounts or early access to promotions, can foster goodwill and encourage continued engagement. These efforts can lead to more detailed and positive reviews, enhancing the business's online reputation and appeal. By leveraging the influence of these users, businesses can position themselves as customer-focused, attract a broader audience, and build stronger loyalty within the community.

### <font face="Helvetica" size=4 color=2E0202> **Q2.2. What are the seasonal and day-of-week trends in restaurant reviews on Yelp?** <font>

```python
%%bigquery
# Seasonality - num of reviews per month
SELECT COUNT(review_id) AS num_reviews, EXTRACT(MONTH FROM date) AS month
FROM `ba775-project-a04.yelp_data.restaurant_review`
GROUP BY month
ORDER BY num_reviews DESC;
```

Our dataset reveals that the highest number of Yelp reviews were posted in July, August, and January (respectively) across all years. This trend likely reflects increased dining activity during these specific months. While there is no consistent seasonality patterns in Yelp review volume, understanding the dynamics of these peaks can help restaurant owners make informed decisions.
January's peak may be tied to the winter break for high school and college students, as well as the New Year holiday period, when people are more likely to travel and dine out. This presents an opportunity for restaurants to target travelers and celebratory diners through special promotions or events. Similarly, July and August correspond to summer vacation, when individuals have more leisure time and disposable income for dining out. Restaurants can capitalize on this by offering summer-themed menus, family dining deals, or outdoor dining experiences to attract more patrons.

The high volume of reviews in these months also suggests increased customer engagement. Restaurants should use this time to amplify their online presence by encouraging satisfied customers to leave reviews and actively responding to feedback to build their reputation. Additionally, operational planning is crucial during these months—ensuring optimal staffing levels, inventory management, and streamlined service to handle increased demand.

Understanding these peak periods allows restaurants to strategically plan marketing campaigns, enhance customer experiences, and maximize profitability during the busiest times for customer interaction on Yelp. Further analysis of the factors driving these trends—such as holidays, school schedules, or tourism patterns—could help refine these strategies even more effectively.



```python
%%bigquery
# Seasonality - num of reviews per day of the week
SELECT
   COUNT(review_id) AS num_reviews,
   FORMAT_DATE('%A', date) AS day_of_week
FROM
   `ba775-project-a04.yelp_data.restaurant_review`
GROUP BY
   day_of_week
ORDER BY
   num_reviews DESC;
```

### <font face="Helvetica" size=4 color=2E0202> **Q2.3. Is review length correlated with rating type, and does a longer review lead to more negativity or usefulness?**<font>

### <font face="Helvetica" size=4 color=2E0202> **Q3.2. What kind of restaurants with high review counts have received compliments on tips?** <font>

```python
%%bigquery
WITH MedianReviewCount AS (
  SELECT
    PERCENTILE_CONT(review_count, 0.5) OVER () AS median_review_count
  FROM
    `ba775-project-a04.yelp_data.restaurant_business`
)

SELECT
#Taking top 15 Chains
  CASE
      WHEN B.name LIKE "%McDonald's%" OR
           B.name LIKE "%Starbucks%" OR
           B.name LIKE "%Chick-fil-A%" OR
           B.name LIKE "%Taco Bell%" OR
           B.name LIKE "%Wendy's%" OR
           B.name LIKE "%Dunkin'%" OR
           B.name LIKE "%Burger King%" OR
           B.name LIKE "%Subway%" OR
           B.name LIKE "%Chipotle Mexican Grill%" OR
           B.name LIKE "%Domino's%" OR
           B.name LIKE "%Panera Bread%" OR
           B.name LIKE "%Panda Express%" OR
           B.name LIKE "%Pizza Hut%" OR
           B.name LIKE "%Sonic Drive-In%" OR
           B.name LIKE "%Popeyes%" OR
           B.name LIKE "%KFC%" OR
           B.name LIKE "%Olive Garden%" OR
           B.name LIKE "%Dairy Queen%" OR
           B.name LIKE "%Texas Roadhouse%" OR
           B.name LIKE "%Arby's%" THEN 'CHAIN'
      ELSE 'INDEPENDENT'
  END AS restaurant_type,
  AVG(B.stars) AS average_stars,
  AVG(compliments_per_business) AS average_received_compliments_count
FROM
  `ba775-project-a04.yelp_data.restaurant_business` AS B
LEFT JOIN (
  SELECT
    T.business_id,
    COUNT(T.compliment_count) AS compliments_per_business
  FROM
    `ba775-project-a04.yelp_data.restaurant_tip` AS T
  GROUP BY
    T.business_id
) AS Compliments
ON
  B.business_id = Compliments.business_id
JOIN
  MedianReviewCount
ON
  TRUE
WHERE
  B.review_count > MedianReviewCount.median_review_count
GROUP BY
  restaurant_type
HAVING
  average_received_compliments_count > 0
ORDER BY
  average_received_compliments_count DESC;
```

The analysis reveals that, on average, independent restaurants (non-franchise businesses) outperform major chain restaurants in both ratings and user engagement, particularly in tips and compliments on tips. Independent businesses receive an average of 22.68 compliments on tips, compared to just 9.63 for chain restaurants.

This significant disparity can be attributed to several factors. First, independent restaurants, being less familiar to tourists and newcomers, may attract more attention on Yelp. Users often rely on reviews and tips for guidance, which increases engagement and the likelihood of compliments. In contrast, chain restaurants, with standardized menus and a more predictable experience, may not inspire the same level of interaction.

Another factor could be the greater emphasis independent businesses place on reputation and customer feedback. These smaller establishments are often more attentive to reviews, responding quickly to negative feedback and addressing areas for improvement. This proactive approach may foster goodwill and lead to more positive reviews and engagement. The stark difference in average ratings—3.81 for independent restaurants versus 2.33 for chains—further supports this explanation, highlighting the potential impact of personalized customer care and service quality on user perceptions.

## <font face="Helvetica" size=5 color=2E0202> **6. Star Ratings and Review Attributes** <font>

### <font face="Helvetica" size=4 color=2E0202> **Q6.1. Is there a relationship between the number of reviews a business has and its star rating?** <font>

```python
%%bigquery
SELECT
   ROUND(AVG(CASE WHEN total_reviews > 2000 THEN average_rating END), 2) AS More_than_2000_reviews,
   ROUND(AVG(CASE WHEN total_reviews <= 2000 AND total_reviews > 1000 THEN average_rating END), 2) AS `1001-2000`,
   ROUND(AVG(CASE WHEN total_reviews <= 1000 AND total_reviews > 500 THEN average_rating END), 2) AS `501-1000`,
   ROUND(AVG(CASE WHEN total_reviews <= 500 AND total_reviews > 200 THEN average_rating END), 2) AS `201-500`,
   ROUND(AVG(CASE WHEN total_reviews <= 200 AND total_reviews > 100 THEN average_rating END), 2) AS `100-200`,
   ROUND(AVG(CASE WHEN total_reviews <= 100 THEN average_rating END), 2) AS Less_than_100_reviews
FROM (
   SELECT
       business_id,
       COUNT(review_id) AS total_reviews,
       AVG(stars) AS average_rating
   FROM
       `ba775-project-a04.yelp_data.restaurant_review`
   GROUP BY
       business_id
);

```

The query explores the relationship between the number of reviews a restaurant receives and its average star rating. It categorizes businesses into review count "buckets," from fewer than 100 reviews to more than 2,000, and calculates the average rating within each category. The results reveal a clear pattern: restaurants with higher review counts tend to have better average ratings. Businesses with over 2,000 reviews achieve an average rating of 4.16, followed closely by those with 1001-2000 reviews at 4.14, and 501-1000 reviews at 3.99. In contrast, restaurants with 201-500 reviews average 3.95, those with 100-200 reviews average 3.85, and those with fewer than 100 reviews drop significantly to 3.33.

This pattern suggests that businesses with more reviews have an established reputation, greater customer trust, and operational consistency, which contribute to higher ratings. Larger businesses also benefit from a diverse customer base that balances occasional negative feedback. In contrast, smaller or newer businesses with fewer reviews are more vulnerable to the impact of outlier ratings. These insights underscore the need for small businesses to focus on increasing review volume to build credibility, while medium and large businesses should prioritize maintaining consistent quality and leveraging their established trust to sustain high ratings.

### <font face="Helvetica" size=4 color=2E0202> **Q6.2. How do star ratings vary with review types (funny, cool, useful) for top-rated vs. low-rated businesses?** <font>

```python
%%bigquery
SELECT
    stars,
    AVG(funny) AS avg_funny,
    AVG(cool) AS avg_cool,
    AVG(useful) AS avg_useful,
    COUNT(*) AS review_count
FROM
    `ba775-project-a04.yelp_data.restaurant_review`
GROUP BY stars
ORDER BY stars;
```

```python
%%bigquery
WITH ReviewCategories AS (
    SELECT
        review_id,
        stars,
        funny,
        cool,
        useful,
        CASE
            WHEN stars >= 4 THEN 'Top Rated'
            WHEN stars <= 2 THEN 'Low Rated'
            ELSE 'Neutral'
        END AS rating_category
    FROM
        `ba775-project-a04.yelp_data.restaurant_review`
)
SELECT
    rating_category,
    AVG(funny) AS avg_funny,
    AVG(cool) AS avg_cool,
    AVG(useful) AS avg_useful,
    COUNT(*) AS review_count
FROM
    ReviewCategories
WHERE
    rating_category IN ('Top Rated', 'Low Rated', 'Neutral')
GROUP BY
    rating_category
ORDER BY
    rating_category;
```

The data highlights clear patterns in how reviews are perceived across star ratings. Neutral reviews (3 stars) are considered the coolest (avg_cool = 0.45), as their balanced and relatable nature resonates with readers. High-rated reviews (4-5 stars), while less funny or useful, stand out as the coolest overall (avg_cool = 0.62) due to their focus on positivity and memorable experiences, though they often lack actionable details.

This dynamic reveals valuable insights for businesses. Low-rated reviews, though critical, should not be overlooked, as they often highlight specific problems that businesses can address to enhance customer satisfaction and prevent future complaints. Engaging with these reviews can also improve a business’s reputation. High-rated reviews underscore the importance of creating positive customer experiences. Encouraging detailed feedback from satisfied customers could make these reviews more actionable, providing prospective customers with a clearer understanding of what to expect.

For Yelp, there’s an opportunity to enhance user engagement by introducing features like a "Funny Reviews" section to showcase humorous low-rated reviews, adding entertainment value to the platform.
